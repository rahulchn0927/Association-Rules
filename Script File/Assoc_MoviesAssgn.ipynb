```python
# conda install -c conda-forge mlxtend
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt
```

# Apriori Algorithm 

```python
# Association rules with 10% Support & 70% Confidence
```

```python
# With 10% Support
frequent_itemsets = apriori(movie1, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
# With 70% Confidence
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
#leverage value : 0 indicates independence. Range will be [-1 1]
# High conviction value means that the consequent is highly depending on the antecedent and range [0 inf]
```

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
# Association rules with 20% Support & 60% Confidence
```

```python
# With 20% Support
frequent_itemsets1 = apriori(movie1, min_support=0.2, use_colnames=True)
frequent_itemsets1
```
